```python
from pathlib import Path

student_id = "2025-0077"
student_name = "KIM HARRIE PILAPIL BEJAR"

# Colab-friendly directory
base_dir = Path("/content/Activity_5_Files")
base_dir.mkdir(parents=True, exist_ok=True)

base_dir
```

```python
file_path = base_dir / f"intro_{student_id}.txt"
file_path.write_text(f"Welcome {student_name} (ID: {student_id}) to File Handling in Python!")

print(f"File created and text written at: {file_path}")
```

```python
content = file_path.read_text()
print(content)
```

```python
with file_path.open("a") as f:
    f.write("\nThis is a new line.")

print(f"Line appended to: {file_path}")
```

```python
file_path2 = base_dir / f"lines_{student_id}.txt"

lines = ["Line 1", "Line 2", "Line 3"]
with file_path2.open("w") as f:
    f.write("\n".join(lines))

print(f"Multiple lines written to: {file_path2}")
```

```python
with file_path2.open("r") as f:
    for line in f:
        print(line.strip())
```

```python
text = file_path2.read_text()
word_count = len(text.split())

print(f"{student_name} (ID: {student_id}) - Word count in file '{file_path2.name}': {word_count}")
```

```python
import shutil

src = file_path
dst = base_dir / f"intro_copy_{student_id}.txt"

shutil.copy(src, dst)
print(f"File copied successfully from {src.name} to {dst.name}.")
```

```python
old_file = dst
new_file = base_dir / f"intro_renamed_{student_id}.txt"

old_file.rename(new_file)
print(f"File renamed successfully from {old_file.name} to {new_file.name}.")
```

```python
if new_file.exists():
    new_file.unlink()
    print(f"File deleted successfully from: {new_file}")
else:
    print("No file found to delete.")
```

```python
new_dir = base_dir / f"data_{student_id}"
new_dir.mkdir(parents=True, exist_ok=True)

print(f"Subdirectory created at: {new_dir}")
```
