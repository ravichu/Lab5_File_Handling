```python
from pathlib import Path
import csv

student_name = "KIM HARRIE PILAPIL BEJAR"
student_id = "2025-0077"

folder = Path("/content/Activity_5_Files")

files = list(folder.glob("*.txt"))
results = []

for file in files:
    text = file.read_text()

    lines_list = [line for line in text.splitlines() if line.strip()]
    num_lines = len(lines_list)

    num_words = len(text.split())
    num_chars = len(text.replace(" ", "").replace("\n", ""))

    wpl = num_words / num_lines if num_lines else 0
    cpw = num_chars / num_words if num_words else 0

    results.append([file.name, num_lines, num_words, num_chars, round(wpl,2), round(cpw,2)])

results.sort(key=lambda x: x[4], reverse=True)

print(f"\nAnalysis by: {student_name} ({student_id})\n")
for r in results:
    print(r)

# Save CSV
csv_file = folder / "content_summary.csv"
with open(csv_file, "w", newline="") as f:
    writer = csv.writer(f)
    writer.writerow(["Filename","Lines","Words","Chars","Words/Line","Chars/Word"])
    writer.writerows(results)

print("Saved:", csv_file)
```

```python
from pathlib import Path
import string

student_name = "KIM HARRIE PILAPIL BEJAR"
student_id = "2025-0077"

folder = Path("/content/Activity_5_Files")

word_freq = {}

for file in folder.glob("*.txt"):
    text = file.read_text().lower()

    for p in string.punctuation:
        text = text.replace(p, "")

    words = text.split()

    for word in words:
        word_freq[word] = word_freq.get(word, 0) + 1

sorted_words = sorted(word_freq.items(), key=lambda x: x[1], reverse=True)

print("\nTop words:")
for w, c in sorted_words[:10]:
    print(w, c)
```

```python
from pathlib import Path
from datetime import datetime
import shutil

student_name = "KIM HARRIE PILAPIL BEJAR"
student_id = "2025-0077"

base_dir = Path("/content/Activity_5_Exercises")
backup_dir = base_dir / f"backup_{student_id}"
backup_dir.mkdir(parents=True, exist_ok=True)

def backup_file(file_path):
    if not file_path.exists():
        print("File not found!")
        return

    timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")

    backup_name = f"{file_path.stem}_{student_id}_{timestamp}{file_path.suffix}"
    backup_path = backup_dir / backup_name

    shutil.copy(file_path, backup_path)

    size = file_path.stat().st_size

    log_file = backup_dir / f"backup_log_{student_id}.txt"

    with open(log_file, "a") as log:
        log.write(f"{student_name} | {student_id}\n")
        log.write(f"{file_path.name} -> {backup_path.name}\n")
        log.write(f"Size: {size} bytes | Time: {timestamp}\n\n")

    print("Backup done:", backup_path)

# create sample file
sample = base_dir / "sample.txt"
base_dir.mkdir(parents=True, exist_ok=True)
sample.write_text("Sample backup file")

backup_file(sample)
```
