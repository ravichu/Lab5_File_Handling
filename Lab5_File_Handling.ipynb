```python
# Step 1: Check if file exists and read whole content
if file_path.exists():
    with open(file_path, "r", encoding="utf-8") as file:
        content = file.read()
    print("File content:\n", content)

# Step 2: Read line by line
with open(file_path, "r", encoding="utf-8") as file:
    for line_number, line in enumerate(file, 1):
        print(f"Line {line_number}: {line.strip()}")
```

```python
# Step 1: Read only lines containing "Python"
with open(file_path, "r", encoding="utf-8") as file:
    for line in file:
        if "Python" in line:
            print(line.strip())

# Step 2: Count words per line
with open(file_path, "r", encoding="utf-8") as file:
    for line in file:
        words = line.split()
        print(f"{line.strip()} -> {len(words)} words")
```

```python
# Step 1: Import libraries
from datetime import datetime
import shutil
from pathlib import Path

# Step 2: Define directory
backup_dir = Path.home() / "Documents" / "PythonFiles"
backup_dir.mkdir(exist_ok=True)

# Step 3: Function to write with backup
def write_with_backup(filename: str, content: str):
    file_path = backup_dir / filename

    # Create backup if file exists
    if file_path.exists():
        timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
        backup_path = file_path.with_name(
            f"{file_path.stem}_backup_{timestamp}{file_path.suffix}"
        )
        shutil.copy2(file_path, backup_path)
        print(f"Backup saved: {backup_path.name}")

    # Write new content
    with open(file_path, "w", encoding="utf-8") as file:
        file.write(content)
        print(f"File saved: {file_path.name}")

# Step 4: Read function
def read_file(filename: str):
    file_path = backup_dir / filename
    with open(file_path, "r", encoding="utf-8") as file:
        return file.read()

# Step 5: Demo
write_with_backup("demo.txt", "Initial content")
write_with_backup("demo.txt", "Updated content")

print(read_file("demo.txt"))
```

```python
# Step 1: Imports
from pathlib import Path
from datetime import datetime
import shutil

# Step 2: Directory
backup_dir = Path.home() / "Documents" / "PythonFiles"
backup_dir.mkdir(exist_ok=True)

# Step 3: File manager
def file_manager():
    file_name = input("Enter filename: ")
    file_path = backup_dir / file_name

    while True:
        print("\n--- MENU ---")
        print("1. Write")
        print("2. Append")
        print("3. Read")
        print("4. Backup")
        print("5. Exit")

        choice = input("Choose: ")

        if choice == "1":
            content = input("Enter content:\n")
            with open(file_path, "w") as f:
                f.write(content)

        elif choice == "2":
            content = input("Append content:\n")
            with open(file_path, "a") as f:
                f.write("\n" + content)

        elif choice == "3":
            if file_path.exists():
                print(file_path.read_text())
            else:
                print("File not found.")

        elif choice == "4":
            if file_path.exists():
                timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
                backup = file_path.with_name(
                    f"{file_path.stem}_backup_{timestamp}{file_path.suffix}"
                )
                shutil.copy2(file_path, backup)
                print("Backup created.")

        elif choice == "5":
            break

# Step 4: Run
file_manager()
```
